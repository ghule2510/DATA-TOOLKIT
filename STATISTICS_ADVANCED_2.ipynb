1.WHAT IS HYPOTHESIS TESTING IN STATISTICS?
ANS-Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on data from a sample.

It helps us decide whether the data provides enough evidence to accept or reject a claim (hypothesis) about the population.

2.WHAT IS THE NULL HYPOTHESIS,AND HOW DOES IT DIFFER FROM THE ALTERNATIVE HYPOTHESIS?
ANS-1. Null Hypothesis (Hâ‚€):

The null hypothesis is a default assumption that there is no effect, no difference, or no relationship in the population.

It acts as the starting point in hypothesis testing.

Example: "The new drug has no effect on blood pressure compared to the old drug."
2. Alternative Hypothesis (Hâ‚ or Ha):

The alternative hypothesis is the statement you want to test/prove.

It suggests that there is an effect, a difference, or a relationship.

Example: "The new drug lowers blood pressure more effectively than the old drug."

6.Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

(Include your Python code and output in the code box below.)
Hint: Generate random number using random function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of random samples

# Generate binomial distribution data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', alpha=0.7, density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)

```python
import numpy as np
from scipy.stats import norm

# Sample dataset
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis setup
mu_0 = 50   # Hypothesized population mean
alpha = 0.05  # significance level

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=0)  # assume population std known
n = len(sample_data)

# Z-statistic calculation
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Critical value for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# P-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Z-statistic:", z_stat)
print("Critical Value:", z_critical)
print("P-value:", p_value)
```

8.Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)
ANS-
1.Simulates data from a normal distribution.

2.Calculates the 95% confidence interval (CI) for the mean.

3.Plots the data using Matplotlib.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu = 50             # True mean
sigma = 5           # True standard deviation
n = 100             # Sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate the 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
confidence_level = 0.95
alpha = 1 - confidence_level
t_crit = stats.t.ppf(1 - alpha/2, df=n-1)  # t-critical value

margin_of_error = t_crit * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 3: Plot the data
plt.figure(figsize=(10,5))
plt.hist(data, bins=15, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Z-score formula: Z = (X - mean) / std
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample standard deviation
    z_scores = (data - mean) / std
    return z_scores

# Simulate some data
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print first 10 Z-scores
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram of Z-scores
plt.figure(figsize=(10,5))
plt.hist(z_scores, bins=15, color='lightcoral', edgecolor='black', alpha=0.7)
plt.axvline(0, color='blue', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
plt.axvline(1, color='green', linestyle='dashed', linewidth=2, label='Z = +1')
plt.axvline(-1, color='green', linestyle='dashed', linewidth=2, label='Z = -1')
plt.title('Histogram of Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
